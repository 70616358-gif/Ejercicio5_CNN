```python
# ===========================================================
# PROYECTO: Tiny ImageNet con Transfer Learning (MobileNetV2)
# METODOLOG√çA: CRISP-DM (Fases 4, 5, y 6)
# ===========================================================
# Curso: Construcci√≥n de Software - Universidad Continental
# ===========================================================

# -----------------------------------------------------------
# PASO 1: IMPORTAR LIBRER√çAS
# -----------------------------------------------------------
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
import os
import datetime
from google.colab import drive, files
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score
import seaborn as sns
from PIL import Image
import io
import zipfile

print("--- Librer√≠as importadas ---")

# -----------------------------------------------------------
# PASO 2: CONECTAR GOOGLE DRIVE
# -----------------------------------------------------------
print("\n--- Conectando a Google Drive ---")
drive.mount('/content/drive')

# -----------------------------------------------------------
# PASO 2.5: DESCARGAR Y PREPARAR DATASET TINY IMAGENET
# -----------------------------------------------------------
print("\n--- Configurando Dataset Tiny ImageNet ---")

# Configurar rutas principales
DRIVE_PATH = "/content/drive/MyDrive/tiny_imagenet_transfer"
DATA_PATH = "/content/tiny-imagenet-200"

os.makedirs(DRIVE_PATH, exist_ok=True)

# Funci√≥n para verificar si el dataset existe
def check_dataset_exists():
    train_path = os.path.join(DATA_PATH, 'train')
    return os.path.exists(train_path) and len(os.listdir(train_path)) > 0

# Descargar dataset si no existe
if not check_dataset_exists():
    print("Dataset no encontrado. Descargando...")

    try:
        # M√©todo 1: Usando kagglehub (m√°s moderno)
        print("Intentando descarga con kagglehub...")
        !pip install -q kagglehub
        import kagglehub

        download_path = kagglehub.dataset_download("akash2sharma/tiny-imagenet")
        print(f"‚úÖ Dataset descargado con kagglehub en: {download_path}")

        # Verificar la estructura y ajustar rutas si es necesario
        if os.path.exists(os.path.join(download_path, 'tiny-imagenet-200')):
            DATA_PATH = os.path.join(download_path, 'tiny-imagenet-200')
        elif os.path.exists(os.path.join(download_path, 'train')):
            DATA_PATH = download_path
        else:
            # Buscar recursivamente la carpeta train
            for root, dirs, files in os.walk(download_path):
                if 'train' in dirs:
                    DATA_PATH = root
                    break

    except Exception as e:
        print(f"‚ùå Error con kagglehub: {e}")
        print("Intentando m√©todo alternativo...")

        try:
            # M√©todo 2: Descarga directa desde un repositorio p√∫blico
            print("Descargando desde repositorio alternativo...")
            !wget -q -O /content/tiny-imagenet-200.zip "https://github.com/arthurdouillard/tiny-imagenet-200/archive/refs/heads/master.zip"

            # Extraer el archivo
            with zipfile.ZipFile('/content/tiny-imagenet-200.zip', 'r') as zip_ref:
                zip_ref.extractall('/content/')

            # Mover y organizar los archivos
            if os.path.exists('/content/tiny-imagenet-200-master'):
                !mv /content/tiny-imagenet-200-master /content/tiny-imagenet-200

            print("‚úÖ Dataset descargado y extra√≠do correctamente")

        except Exception as e2:
            print(f"‚ùå Error con descarga alternativa: {e2}")
            print("\nüìù SOLUCI√ìN MANUAL REQUERIDA:")
            print("1. Ve a: https://www.kaggle.com/datasets/akash2sharma/tiny-imagenet")
            print("2. Haz clic en 'Download' (necesitar√°s cuenta de Kaggle)")
            print("3. Sube el archivo ZIP manualmente a Google Drive")
            print("4. Modifica la variable DATA_PATH para apuntar a tu archivo")
            raise Exception("No se pudo descargar el dataset autom√°ticamente")

else:
    print("‚úÖ Dataset ya existe")

print(f"üìÅ Ruta final del dataset: {DATA_PATH}")

# -----------------------------------------------------------
# PASO 3: CONFIGURAR RUTAS Y PAR√ÅMETROS RESTANTES
# -----------------------------------------------------------
print("\n--- Configurando Rutas y Par√°metros ---")

MODEL_PATH = os.path.join(DRIVE_PATH, "mobilenetv2_tiny_imagenet_model.keras")
LOG_PATH = os.path.join(DRIVE_PATH, "logs")

os.makedirs(LOG_PATH, exist_ok=True)

IMG_SIZE = (96, 96)  # Se usa 96x96 para MobileNetV2
BATCH_SIZE = 32
EPOCHS = 10
NUM_CLASSES = 200

print(f"Modelo se guardar√° en: {MODEL_PATH}")
print(f"Im√°genes de tama√±o: {IMG_SIZE}")
print(f"Dataset en: {DATA_PATH}")

# Verificaci√≥n final de la estructura del dataset
print("\n--- Verificando estructura del dataset ---")
train_dir = os.path.join(DATA_PATH, 'train')
if os.path.exists(train_dir):
    num_classes = len([name for name in os.listdir(train_dir) if os.path.isdir(os.path.join(train_dir, name))])
    print(f"‚úÖ Encontradas {num_classes} clases en el directorio de entrenamiento")
else:
    print("‚ùå No se encuentra el directorio 'train'")
    print("Contenido del directorio principal:")
    print(os.listdir(DATA_PATH))
    raise FileNotFoundError(f"No se encontr√≥ el directorio 'train' en {DATA_PATH}")

# -----------------------------------------------------------
# PASO 4: CARGAR Y PREPARAR DATOS
# -----------------------------------------------------------
print("\n--- Preparando generadores de datos ---")

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    validation_split=0.1
)

train_gen = train_datagen.flow_from_directory(
    os.path.join(DATA_PATH, 'train'),
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    shuffle=True
)

val_gen = train_datagen.flow_from_directory(
    os.path.join(DATA_PATH, 'train'),
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    shuffle=False  # Important para evaluaci√≥n consistente
)

print(f"‚úÖ Clases encontradas: {len(train_gen.class_indices)}")

# -----------------------------------------------------------
# DICCIONARIO DE NOMBRES LEGIBLES PARA LAS CLASES
# -----------------------------------------------------------
# Mapeo de IDs de ImageNet a nombres legibles en espa√±ol/ingl√©s
IMAGENET_ID_TO_NAME = {
    'n01443537': 'goldfish (pez dorado)',
    'n01629819': 'European fire salamander (salamandra)',
    'n01641577': 'bullfrog (rana toro)',
    'n01644900': 'tailed frog (rana con cola)',
    'n01698640': 'American alligator (caim√°n)',
    'n01742172': 'boa constrictor (boa)',
    'n01768244': 'trilobite (trilobites)',
    'n01770393': 'scorpion (escorpi√≥n)',
    'n01774384': 'black widow (ara√±a viuda negra)',
    'n01774750': 'tarantula (tar√°ntula)',
    'n01784675': 'centipede (ciempi√©s)',
    'n01855672': 'flamingo (flamenco)',
    'n01882714': 'kiwi (kiwi)',
    'n01910747': 'jellyfish (medusa)',
    'n01917289': 'brain coral (coral cerebro)',
    'n01944390': 'snail (caracol)',
    'n01945685': 'slug (babosa)',
    'n01950731': 'sea anemone (an√©mona de mar)',
    'n01983481': 'american lobster (langosta americana)',
    'n01984695': 'spiny lobster (langosta espinosa)',
    'n02002724': 'great white shark (tibur√≥n blanco)',
    'n02056570': 'king penguin (ping√ºino rey)',
    'n02058221': 'albatross (albatros)',
    'n02074367': 'dugong (dug√≥n)',
    'n02085620': 'Chihuahua (chihuahua)',
    'n02094433': 'Yorkshire terrier (yorkshire terrier)',
    'n02099601': 'golden retriever (golden retriever)',
    'n02099712': 'Labrador retriever (labrador retriever)',
    'n02106662': 'German shepherd (pastor alem√°n)',
    'n02113799': 'standard poodle (caniche est√°ndar)',
    'n02123045': 'tabby cat (gato atigrado)',
    'n02123394': 'Persian cat (gato persa)',
    'n02124075': 'Egyptian cat (gato egipcio)',
    'n02125311': 'cougar (puma)',
    'n02129165': 'lion (le√≥n)',
    'n02132136': 'brown bear (oso pardo)',
    'n02165456': 'ladybug (mariquita)',
    'n02190166': 'fly (mosca)',
    'n02206856': 'bee (abeja)',
    'n02226429': 'grasshopper (saltamontes)',
    'n02231487': 'walking stick (insecto palo)',
    'n02233338': 'cockroach (cucaracha)',
    'n02236044': 'mantis (mantis religiosa)',
    'n02268443': 'dragonfly (lib√©lula)',
    'n02279972': 'monarch butterfly (mariposa monarca)',
    'n02281406': 'sulphur butterfly (mariposa azufre)',
    'n02321529': 'sea urchin (erizo de mar)',
    'n02364673': 'guinea pig (conejillo de indias)',
    'n02395406': 'hog (cerdo)',
    'n02403003': 'ox (buey)',
    'n02410509': 'bison (bisonte)',
    'n02415577': 'bighorn sheep (carnero de las rocosas)',
    'n02423022': 'gazelle (gacela)',
    'n02437312': 'camel (camello)',
    'n02480495': 'orangutan (orangut√°n)',
    'n02481823': 'chimpanzee (chimpanc√©)',
    'n02486410': 'baboon (babuino)',
    'n02504458': 'elephant (elefante)',
    'n02509815': 'red panda (panda rojo)',
    'n02669723': 'academic gown (toga acad√©mica)',
    'n02769748': 'backpack (mochila)',
    'n02788148': 'barbell (pesas)',
    'n02791270': 'barbershop (barber√≠a)',
    'n02795169': 'beer bottle (botella de cerveza)',
    'n02802426': 'basketball (baloncesto)',
    'n02808440': 'bathtub (ba√±era)',
    'n02814860': 'beacon (faro)',
    'n02815834': 'beaker (vaso de precipitados)',
    'n02823428': 'beer glass (vaso de cerveza)',
    'n02841315': 'binoculars (prism√°ticos)',
    'n02843684': 'birdhouse (casa de p√°jaros)',
    'n02883205': 'bow tie (corbat√≠n)',
    'n02892201': 'brassiere (sost√©n)',
    'n02906734': 'broom (escoba)',
    'n02927161': 'butcher shop (carnicer√≠a)',
    'n02948072': 'candle (vela)',
    'n02950826': 'cannon (ca√±√≥n)',
    'n02963159': 'cardigan (c√°rdigan)',
    'n02999410': 'chain (cadena)',
    'n03014705': 'chest (cofre)',
    'n03042490': 'cliff dwelling (vivienda en acantilado)',
    'n03085013': 'computer keyboard (teclado)',
    'n03089624': 'confectionery (confiter√≠a)',
    'n03126707': 'crane (gr√∫a)',
    'n03160309': 'dam (presa)',
    'n03201208': 'dining table (mesa de comedor)',
    'n03255030': 'dumbbell (mancuerna)',
    'n03355925': 'flag (bandera)',
    'n03388043': 'fountain (fuente)',
    'n03393912': 'freight car (vag√≥n de carga)',
    'n03400231': 'frying pan (sart√©n)',
    'n03404251': 'fur coat (abrigo de piel)',
    'n03424325': 'gasmask (m√°scara de gas)',
    'n03444034': 'go-kart (kart)',
    'n03447447': 'gondola (g√≥ndola)',
    'n03544143': 'hourglass (reloj de arena)',
    'n03617480': 'kimono (quimono)',
    'n03637318': 'lampshade (pantalla de l√°mpara)',
    'n03649909': 'lawn mower (cortadora de c√©sped)',
    'n03662601': 'lifeboat (bote salvavidas)',
    'n03706229': 'magnetic compass (br√∫jula)',
    'n03763968': 'military uniform (uniforme militar)',
    'n03770439': 'miniskirt (minifalda)',
    'n03804744': 'nail (clavo)',
    'n03838899': 'oboe (oboe)',
    'n03891332': 'parking meter (parqu√≠metro)',
    'n03902125': 'payphone (tel√©fono p√∫blico)',
    'n03930313': 'picket fence (valla de estacas)',
    'n03937543': 'pill bottle (frasco de p√≠ldoras)',
    'n03976657': 'pole (poste)',
    'n03992509': 'potter\'s wheel (torno de alfarero)',
    'n04023962': 'punching bag (saco de boxeo)',
    'n04070727': 'radio telescope (radiotelescopio)',
    'n04099969': 'rocking chair (mecedora)',
    'n04118538': 'rugby ball (bal√≥n de rugby)',
    'n04133789': 'sandal (sandalia)',
    'n04149813': 'saxophone (saxof√≥n)',
    'n04179913': 'sewing machine (m√°quina de coser)',
    'n04251144': 'snorkel (esn√≥rquel)',
    'n04254777': 'sock (calcet√≠n)',
    'n04259630': 'solar dish (plato solar)',
    'n04265275': 'space heater (calentador espacial)',
    'n04275548': 'spider web (telara√±a)',
    'n04285008': 'sports car (coche deportivo)',
    'n04311004': 'steel arch bridge (puente de arco de acero)',
    'n04328186': 'stove (estufa)',
    'n04356056': 'sunglasses (gafas de sol)',
    'n04366367': 'suspension bridge (puente colgante)',
    'n04371430': 'swimsuit (traje de ba√±o)',
    'n04398044': 'teapot (tetera)',
    'n04399382': 'teddy bear (oso de peluche)',
    'n04417672': 'thatch (paja)',
    'n04456115': 'tollbooth (cabina de peaje)',
    'n04465501': 'tractor (tractor)',
    'n04487081': 'trolleybus (troleb√∫s)',
    'n04532106': 'vase (jarr√≥n)',
    'n04532670': 'viaduct (viaducto)',
    'n04540053': 'volleyball (voleibol)',
    'n04560804': 'water jug (jarra de agua)',
    'n04562935': 'water tower (torre de agua)',
    'n04596742': 'wok (wok)',
    'n07579787': 'plate (plato)',
    'n07583066': 'guacamole (guacamole)',
    'n07614500': 'ice cream (helado)',
    'n07615774': 'ice lolly (polo de hielo)',
    'n07695742': 'pretzel (pretzel)',
    'n07711569': 'mashed potato (pur√© de patatas)',
    'n07715103': 'cauliflower (coliflor)',
    'n07720875': 'bell pepper (pimiento)',
    'n07734744': 'mushroom (champi√±√≥n)',
    'n07747607': 'orange (naranja)',
    'n07749582': 'lemon (lim√≥n)',
    'n07753592': 'banana (pl√°tano)',
    'n07768694': 'pomegranate (granada)',
    'n07871810': 'meat loaf (pastel de carne)',
    'n07873807': 'pizza (pizza)',
    'n07875152': 'potpie (pastel de carne)',
    'n09246464': 'cliff (acantilado)',
    'n09256479': 'coral reef (arrecife de coral)',
    'n12267677': 'acorn (bellota)'
}

print(f"‚úÖ Diccionario de nombres cargado con {len(IMAGENET_ID_TO_NAME)} entradas")

# -----------------------------------------------------------
# PASO 5: DEFINIR MODELO CON TRANSFER LEARNING (FASE 4)
# -----------------------------------------------------------
print("\n--- FASE 4: MODELADO (Transfer Learning con MobileNetV2) ---")

# 1Ô∏è‚É£ Cargar modelo base preentrenado
base_model = keras.applications.MobileNetV2(
    weights='imagenet',
    include_top=False,
    input_shape=IMG_SIZE + (3,)
)

# 2Ô∏è‚É£ Congelar capas base (para no reentrenarlas)
base_model.trainable = False

# 3Ô∏è‚É£ A√±adir capas personalizadas
model = keras.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(512, activation='relu'),
    layers.Dropout(0.4),
    layers.Dense(NUM_CLASSES, activation='softmax')
])

# 4Ô∏è‚É£ Compilar modelo
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.0005),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# -----------------------------------------------------------
# PASO 6: CONFIGURAR CALLBACKS Y ENTRENAR MODELO
# -----------------------------------------------------------
print("\n--- Entrenando modelo (fase 4) ---")

# Calcular steps_per_epoch correctamente
steps_per_epoch = max(1, train_gen.samples // BATCH_SIZE)
validation_steps = max(1, val_gen.samples // BATCH_SIZE)

print(f"Steps por √©poca: {steps_per_epoch}")
print(f"Validation steps: {validation_steps}")

log_dir = os.path.join(LOG_PATH, datetime.datetime.now().strftime("%Y%m%d-%H%M%S"))
callbacks = [
    tf.keras.callbacks.TensorBoard(log_dir=log_dir),
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-7)
]

history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=EPOCHS,
    callbacks=callbacks,
    steps_per_epoch=steps_per_epoch,
    validation_steps=validation_steps,
    verbose=1
)

model.save(MODEL_PATH)
print(f"‚úÖ Modelo guardado en: {MODEL_PATH}")

# Gr√°ficas de entrenamiento
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Validaci√≥n')
plt.title('Precisi√≥n del Modelo')
plt.xlabel('√âpoca')
plt.ylabel('Precisi√≥n')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Entrenamiento')
plt.plot(history.history['val_loss'], label='Validaci√≥n')
plt.title('P√©rdida del Modelo')
plt.xlabel('√âpoca')
plt.ylabel('P√©rdida')
plt.legend()

plt.tight_layout()
plt.show()

# -----------------------------------------------------------
# PASO 7: FASE 5 ‚Äì EVALUACI√ìN COMPLETA CON NOMBRES LEGIBLES
# -----------------------------------------------------------
print("\n" + "="*60)
print("FASE 5: EVALUACI√ìN COMPLETA DEL MODELO")
print("="*60)

# Evaluaci√≥n b√°sica con TODOS los datos de validaci√≥n
print("Evaluando con todos los datos de validaci√≥n...")
val_loss, val_acc = model.evaluate(val_gen, verbose=1)
print(f"P√©rdida en validaci√≥n: {val_loss:.4f}")
print(f"Precisi√≥n en validaci√≥n: {val_acc*100:.2f}%")

# Obtener predicciones de TODOS los datos de validaci√≥n
print("\n--- Obteniendo predicciones completas ---")

# Reiniciar el generador y obtener TODAS las muestras
val_gen.reset()
y_true_all = []
y_pred_all = []

# Procesar TODOS los batches de validaci√≥n
total_batches = val_gen.samples // BATCH_SIZE + (1 if val_gen.samples % BATCH_SIZE != 0 else 0)

print(f"Procesando {total_batches} batches completos...")
for i in range(total_batches):
    if i % 20 == 0:
        print(f"Procesando batch {i}/{total_batches}")
    try:
        x_val, y_val = next(val_gen)
        y_pred_probs = model.predict(x_val, verbose=0)
        y_pred_batch = np.argmax(y_pred_probs, axis=1)
        y_true_batch = np.argmax(y_val, axis=1)

        y_true_all.extend(y_true_batch)
        y_pred_all.extend(y_pred_batch)
    except StopIteration:
        break

y_true_all = np.array(y_true_all)
y_pred_all = np.array(y_pred_all)

print(f"‚úÖ Predicciones completas obtenidas: {len(y_true_all)} muestras")

# Verificar que tenemos las 200 clases
unique_classes_true = np.unique(y_true_all)
unique_classes_pred = np.unique(y_pred_all)
unique_classes = np.unique(np.concatenate([y_true_all, y_pred_all]))

print(f"Clases √∫nicas en etiquetas reales: {len(unique_classes_true)}")
print(f"Clases √∫nicas en predicciones: {len(unique_classes_pred)}")
print(f"Clases √∫nicas totales: {len(unique_classes)}")

# Distribuci√≥n de clases en la evaluaci√≥n
print("\n--- Distribuci√≥n de Clases en la Evaluaci√≥n ---")
class_distribution = {}
for class_idx in range(NUM_CLASSES):
    count = np.sum(y_true_all == class_idx)
    class_distribution[class_idx] = count

# Mostrar clases con 0 muestras
zero_sample_classes = [cls for cls, count in class_distribution.items() if count == 0]
if zero_sample_classes:
    print(f"‚ö†Ô∏è  Clases con 0 muestras en validaci√≥n: {len(zero_sample_classes)}")
else:
    print("‚úÖ Todas las clases tienen muestras en validaci√≥n")

# Mostrar distribuci√≥n
plt.figure(figsize=(15, 6))
counts = [class_distribution[i] for i in range(NUM_CLASSES)]
plt.bar(range(NUM_CLASSES), counts, color='lightblue', alpha=0.7)
plt.axhline(y=np.mean(counts), color='red', linestyle='--', label=f'Promedio: {np.mean(counts):.1f}')
plt.xlabel('Clases')
plt.ylabel('N√∫mero de Muestras')
plt.title('Distribuci√≥n de Muestras por Clase en Validaci√≥n')
plt.legend()
plt.tight_layout()
plt.show()

# Reporte de Clasificaci√≥n COMPLETO con nombres legibles
print("\n--- Reporte de Clasificaci√≥n COMPLETO (200 clases) ---")

class_names = list(val_gen.class_indices.keys())

# Crear nombres legibles para todas las clases
readable_names = []
for class_id in class_names:
    if class_id in IMAGENET_ID_TO_NAME:
        readable_names.append(f"{class_id}\n({IMAGENET_ID_TO_NAME[class_id]})")
    else:
        readable_names.append(f"{class_id}\n(desconocido)")

# Reporte para todas las clases
all_classes = list(range(NUM_CLASSES))

print(classification_report(
    y_true_all,
    y_pred_all,
    labels=all_classes,
    target_names=readable_names,
    zero_division=0,
    digits=4
))

# M√©tricas globales
accuracy = accuracy_score(y_true_all, y_pred_all)
precision = precision_score(y_true_all, y_pred_all, average='macro', zero_division=0)
recall = recall_score(y_true_all, y_pred_all, average='macro', zero_division=0)

print(f"\n--- M√©tricas Globales ---")
print(f"Exactitud (Accuracy): {accuracy*100:.2f}%")
print(f"Precisi√≥n (Precision macro): {precision*100:.2f}%")
print(f"Sensibilidad (Recall macro): {recall*100:.2f}%")

# Matriz de Confusi√≥n para las primeras 20 clases
print("\n--- Matriz de Confusi√≥n (primeras 20 clases) ---")

top_classes = list(range(min(20, NUM_CLASSES)))
mask = np.isin(y_true_all, top_classes) & np.isin(y_pred_all, top_classes)

if np.sum(mask) > 0:
    y_true_filtered = y_true_all[mask]
    y_pred_filtered = y_pred_all[mask]

    # Mapear a √≠ndices 0-19 para mejor visualizaci√≥n
    class_mapping = {cls: i for i, cls in enumerate(top_classes)}
    y_true_mapped = np.array([class_mapping[cls] for cls in y_true_filtered])
    y_pred_mapped = np.array([class_mapping[cls] for cls in y_pred_filtered])

    class_names_filtered = [readable_names[i] for i in top_classes]

    plt.figure(figsize=(16, 14))
    cm = confusion_matrix(y_true_mapped, y_pred_mapped)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names_filtered,
                yticklabels=class_names_filtered)
    plt.title("Matriz de Confusi√≥n - Tiny ImageNet (primeras 20 clases)")
    plt.xlabel("Predicci√≥n")
    plt.ylabel("Real")
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
else:
    print("No hay suficientes datos para generar la matriz de confusi√≥n")

# Gr√°fico de precisi√≥n por clase (Top 25)
print("\n--- Precisi√≥n por Clase (Top 25) ---")
class_accuracy = {}
for class_idx in range(NUM_CLASSES):
    mask = y_true_all == class_idx
    if np.sum(mask) > 0:
        class_acc = np.mean(y_pred_all[mask] == class_idx)
        class_accuracy[class_idx] = (class_acc, np.sum(mask))

# Ordenar por precisi√≥n
sorted_classes = sorted(class_accuracy.items(), key=lambda x: x[1][0], reverse=True)[:25]

plt.figure(figsize=(14, 10))
classes_plot = [f"{readable_names[cls]}\n(n={count})" for cls, (acc, count) in sorted_classes]
accuracies_plot = [acc * 100 for cls, (acc, count) in sorted_classes]

plt.barh(range(len(classes_plot)), accuracies_plot, color='skyblue')
plt.xlabel('Precisi√≥n (%)')
plt.title('Top 25 Clases con Mejor Precisi√≥n')
plt.yticks(range(len(classes_plot)), classes_plot, fontsize=8)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# -----------------------------------------------------------
# PASO 8: FASE 6 ‚Äì DESPLIEGUE INTERACTIVO CON FEEDBACK MEJORADO
# -----------------------------------------------------------
print("\n" + "="*60)
print("FASE 6: DESPLIEGUE INTERACTIVO CON FEEDBACK")
print("="*60)

def predict_custom_image():
    print("\n--- Prueba con imagen personalizada ---")
    print("Por favor, sube una imagen para clasificar...")

    uploaded = files.upload()
    if not uploaded:
        print("No se subi√≥ ninguna imagen.")
        return

    filename = next(iter(uploaded))

    try:
        # Procesar imagen
        img = Image.open(io.BytesIO(uploaded[filename]))
        img = img.convert('RGB')
        img_resized = img.resize(IMG_SIZE, Image.Resampling.LANCZOS)

        # Preparar para predicci√≥n
        img_array = np.array(img_resized).astype("float32") / 255.0
        img_batch = np.expand_dims(img_array, axis=0)

        # Realizar predicci√≥n
        prediction = model.predict(img_batch, verbose=0)
        predicted_label = np.argmax(prediction[0])
        confidence = np.max(prediction[0]) * 100

        # Obtener top 5 predicciones
        top_5_indices = np.argsort(prediction[0])[-5:][::-1]
        top_5_confidences = prediction[0][top_5_indices] * 100

        # Mapeo de clases con nombres legibles
        idx_to_class = {v: k for k, v in val_gen.class_indices.items()}
        predicted_class_id = idx_to_class.get(predicted_label, "Desconocida")
        predicted_class_name = IMAGENET_ID_TO_NAME.get(predicted_class_id, "Clase desconocida")

        # Mostrar resultados
        plt.figure(figsize=(12, 6))

        plt.subplot(1, 2, 1)
        plt.imshow(img_resized)
        plt.title(f"Imagen Original\n{filename}")
        plt.axis('off')

        plt.subplot(1, 2, 2)
        # Mostrar top 5 predicciones con nombres legibles
        top_class_ids = [idx_to_class.get(i, "Desconocida") for i in top_5_indices]
        top_class_names = [IMAGENET_ID_TO_NAME.get(cls_id, "Clase desconocida") for cls_id in top_class_ids]
        colors = plt.cm.RdYlGn(np.linspace(0.3, 0.7, 5))

        bars = plt.barh(range(5), top_5_confidences[::-1], color=colors[::-1])
        plt.yticks(range(5), [f"{name}" for name in top_class_names[::-1]], fontsize=9)
        plt.xlabel('Confianza (%)')
        plt.title('Top 5 Predicciones')

        # A√±adir valores en las barras
        for i, (bar, conf) in enumerate(zip(bars, top_5_confidences[::-1])):
            plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2,
                    f'{conf:.1f}%', va='center', fontsize=10)

        plt.tight_layout()
        plt.show()

        print(f"\nüîç RESULTADO PRINCIPAL:")
        print(f"   ID: {predicted_class_id}")
        print(f"   Clase: {predicted_class_name}")
        print(f"   Confianza: {confidence:.2f}%")

        print(f"\nüìä TOP 5 PREDICCIONES:")
        for i, (cls_id, cls_name, conf) in enumerate(zip(top_class_ids, top_class_names, top_5_confidences)):
            print(f"   {i+1}. {cls_name} ({cls_id}): {conf:.2f}%")

    except Exception as e:
        print(f"‚ùå Error al procesar la imagen: {e}")

# Ejecutar la predicci√≥n interactiva
predict_custom_image()

print("\n" + "="*60)
print("--- PROYECTO COMPLETADO ---")
print("="*60)
print("Fase 4 (Modelado) ‚úÖ - Transfer Learning con MobileNetV2")
print("Fase 5 (Evaluaci√≥n) ‚úÖ - Evaluaci√≥n completa con nombres legibles")
print("Fase 6 (Despliegue) ‚úÖ - Interacci√≥n mejorada con nombres")
print(f"Clases evaluadas: {len(unique_classes_true)}/200")
print(f"Precisi√≥n final: {val_acc*100:.2f}%")
print("\nüéâ ¬°Proyecto ejecutado exitosamente!")
```
